## Частина перша: Знайомство з Pandas.

### Вивести перші рядки таблиці за допомогою методу head

```python
url = "https://uk.wikipedia.org/wiki/%D0%9D%D0%B0%D1%80%D0%BE%D0%B4%D0%B6%D1%83%D0%B2%D0%B0%D0%BD%D1%96%D1%81%D1%82%D1%8C_%D0%B2_%D0%A3%D0%BA%D1%80%D0%B0%D1%97%D0%BD%D1%96"

all_tables = pd.read_html(url,
                          thousands="",
                          decimal=",",
                          match = "Коефіцієнт народжуваності у регіонах України у 1950—2014 роках",
                    )
main_table = pd.DataFrame(all_tables[0])
main_table.head()
```

### Визначте кількість рядків та стовпців у датафреймі (атрибут shape)

```python
main_table.shape
```

### Замініть у таблиці значення "—" на значення NaN

```python
table1 = main_table
table1.replace('—', pd.NA, inplace=True)
table1
```

### Визначте типи всіх стовпців за допомогою dataframe.dtypes

```python
main_table.dtypes
```

### Замініть типи нечислових колонок на числові. Підказка - це колонки, де знаходився символ "—"

```python
table2 = table1
table2.fillna(0, inplace=True)
table2
```

### Порахуйте, яка частка пропусків міститься в кожній колонці (використовуйте методи isnull та sum)

```python
table_missed = main_table.isnull().sum(axis=0) / len(main_table)
table_missed
```

### Видаліть з таблиці дані по всій країні, останній рядок таблиці

```python
main_table = main_table.drop(main_table.index[len((main_table))-1])
main_table
```

### Замініть відсутні дані в стовпцях середніми значеннями цих стовпців (метод fillna)

```python
means = main_table.mean(axis=0, numeric_only=True)
print(means)
table4 = main_table.fillna(means)
table4
```

### Отримайте список регіонів, де рівень народжуваності у 2019 році був вищим за середній по Україні

```python
means2014 = main_table.mean(axis=0, numeric_only=True)['2014']
print(means2014)
main_table[main_table['2014'] > means2014]
```

### У якому регіоні була найвища народжуваність у 2014 році?

```python

main_table[main_table['2014'] == main_table['2014'].max()]
```

### Побудуйте стовпчикову діаграму народжуваності по регіонах у 2019 році

```python
plt.bar(table4['регіон'],table4['2014'], color='g')
plt.xticks(rotation=90)
plt.ylabel('Коеф. народжуваності')
plt.show()
```
